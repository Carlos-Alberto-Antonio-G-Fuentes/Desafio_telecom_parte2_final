##An√°lisis de multicolinealidad con VIF sobre variables num√©ricas

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
import pandas as pd

# Crear un DataFrame solo con las variables explicativas (excluyendo la variable objetivo 'churn')
X_num = datos_codificados.drop(columns=["Churn"])[['MonthlyCharges', 'TotalCharges', 'Cuentas_Diarias', 'tenure']]
X_num = add_constant(X_num)

# Calcular VIF
vif_data = pd.DataFrame()
vif_data["Variable"] = X_num.columns
vif_data["VIF"] = [variance_inflation_factor(X_num.values, i) for i in range(X_num.shape[1])]

# Mostrar resultados ordenados
print(vif_data.sort_values(by="VIF", ascending=False))
```

**Interpretaci√≥n del VIF**

El VIF (Variance Inflation Factor) nos dice cu√°nto se multiplica la varianza de una variable explicativa por la multicolinealidad. Idealmente, los valores deber√≠an estar por debajo de 5, aunque algunos toleran hasta 10. Multicolinealidad extrema (inf): probablemente es combinaci√≥n lineal exacta con otra variable, por lo cual solo dejaremos las variables MonthlyCharges y tenure y eliminaremos las variables Cuentas_Diarias y TotalCharges por redundancia en nuestro modelo.

Const es el valor, de la constante del modelo.

##Graficando variables num√©ricas

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches  # Para crear la leyenda manualmente

# Variables num√©ricas a graficar
variables =variables_numericas

# Paleta de colores personalizada
colores = {0: "#1f77b4", 1: "#ff7f0e"}  # Azul para No Churn, Naranjo para Churn

# Crear figura y ejes
fig, axs = plt.subplots(1, 2, figsize=(10,6))
axs = axs.flatten()

for ax, variable in zip(axs, variables):
    # Boxplot con color y media
    sns.boxplot(
        data=datos_depurados,
        x='Churn',
        y=variable,
        ax=ax,
        hue='Churn',
        palette=colores,
        dodge=False,
        showmeans=True,
        meanline=True,
        meanprops={"color": "red", "linestyle": "--", "linewidth": 2},
        legend=False  # Evita duplicaci√≥n de leyenda
    )

    # Calcular medianas
    mediana_0 = datos_depurados[datos_depurados['Churn'] == 0][variable].median()
    mediana_1 = datos_depurados[datos_depurados['Churn'] == 1][variable].median()

    # A√±adir texto de mediana
    ax.text(0, mediana_0, f'Mediana: {mediana_0:.2f}', ha='center', va='bottom', fontsize=10, color='black')
    ax.text(1, mediana_1, f'Mediana: {mediana_1:.2f}', ha='center', va='bottom', fontsize=10, color='black')

    # T√≠tulo y etiquetas
    ax.set_xticklabels(['No Churn', 'Churn'])
    ax.set_title(f'Distribuci√≥n de {variable} por Churn')

# Agregar leyenda manual (fuera del loop)
leyenda_patches = [
    mpatches.Patch(color=colores[0], label='No Churn (0)'),
    mpatches.Patch(color=colores[1], label='Churn (1)')
]

# Posicionar leyenda arriba a la derecha
fig.legend(
    handles=leyenda_patches,
    loc='upper right',
    title='Churn',
    fontsize=6,
    title_fontsize=6
)

# Mostrar y guardar
plt.tight_layout(rect=[0, 0, 0.95, 0.95])  # Ajustar espacio para la leyenda
plt.savefig("3_graficos_boxplot_variables_numericas.png", dpi=300)
plt.show()
```

##Interpretaci√≥n del Mapa de Calor de correlaciones con Churn

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# Calculamos la correlaci√≥n
correlaciones = datos_depurados.corr(numeric_only=True)

# Extraemos la fila de 'Churn' (sin incluir 'Churn' consigo mismo)
corr_churn = correlaciones.loc['Churn'].drop('Churn')

# Ordenamos por magnitud (valor absoluto)
corr_ordenadas = corr_churn.reindex(corr_churn.abs().sort_values(ascending=False).index)

# Clasificaci√≥n del tipo de correlaci√≥n
def clasificar_correlacion(valor):
    abs_val = abs(valor)
    if abs_val >= 0.3:
        return 'Alta'
    elif abs_val >= 0.2:
        return 'Media'
    elif abs_val >= 0.1:
        return 'Baja'
    else:
        return 'Insignificante'

# Interpretaci√≥n contextual
def interpretar_correlacion(var, val):
    if val > 0:
        return f'A mayor {var}, mayor probabilidad de churn'
    elif val < 0:
        return f'A mayor {var}, menor probabilidad de churn'
    else:
        return f'{var} no tiene relaci√≥n con churn'

# Construimos el DataFrame
tabla_corr = pd.DataFrame({
    'Variable': corr_ordenadas.index,
    'Correlaci√≥n': corr_ordenadas.values,
    'Signo': ['+' if val > 0 else '‚Äì' for val in corr_ordenadas.values],
    'Tipo de Correlaci√≥n': [clasificar_correlacion(val) for val in corr_ordenadas.values],
    'Interpretaci√≥n': [interpretar_correlacion(var, val) for var, val in zip(corr_ordenadas.index, corr_ordenadas.values)]
})

# Crear figura tipo tabla con Plotly
fig = go.Figure(data=[go.Table(
    columnwidth=[70, 40, 40, 70, 200],
    header=dict(
        values=[
            'Variable',
            'Correlaci√≥n<br><span style="font-size:11px">',
            'Signo<br><span style="font-size:11px">',
            'Tipo de Correlaci√≥n<br><span style="font-size:11px">',
            'Interpretaci√≥n<br><span style="font-size:11px">'
        ],
        fill_color='lightsteelblue',
        align='center',
        font=dict(size=14, color='black'),
        height=40
    ),
    cells=dict(
        values=[
            tabla_corr['Variable'],
            [f"{val:.2f}" for val in tabla_corr['Correlaci√≥n']],
            tabla_corr['Signo'],
            tabla_corr['Tipo de Correlaci√≥n'],
            tabla_corr['Interpretaci√≥n']
        ],
        fill_color='white',
        align='center',
        font=dict(size=13),
        height=30
    )
)])

fig.update_layout(margin=dict(l=40, r=40, t=20, b=20))
fig.write_html("7_Interpretacion_Correlaciones_Churn.html")
fig.show()




```
