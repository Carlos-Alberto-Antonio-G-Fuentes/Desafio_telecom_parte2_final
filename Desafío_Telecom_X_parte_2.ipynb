##Análisis de multicolinealidad con VIF sobre variables numéricas

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
import pandas as pd

# Crear un DataFrame solo con las variables explicativas (excluyendo la variable objetivo 'churn')
X_num = datos_codificados.drop(columns=["Churn"])[['MonthlyCharges', 'TotalCharges', 'Cuentas_Diarias', 'tenure']]
X_num = add_constant(X_num)

# Calcular VIF
vif_data = pd.DataFrame()
vif_data["Variable"] = X_num.columns
vif_data["VIF"] = [variance_inflation_factor(X_num.values, i) for i in range(X_num.shape[1])]

# Mostrar resultados ordenados
print(vif_data.sort_values(by="VIF", ascending=False))
```

**Interpretación del VIF**

El VIF (Variance Inflation Factor) nos dice cuánto se multiplica la varianza de una variable explicativa por la multicolinealidad. Idealmente, los valores deberían estar por debajo de 5, aunque algunos toleran hasta 10. Multicolinealidad extrema (inf): probablemente es combinación lineal exacta con otra variable, por lo cual solo dejaremos las variables MonthlyCharges y tenure y eliminaremos las variables Cuentas_Diarias y TotalCharges por redundancia en nuestro modelo.

Const es el valor, de la constante del modelo.

* OnlineSecurity / OnlineBackup / TechSupport / DeviceProtection: Los que no usan servicios adicionales de soporte o seguridad tienen tasas de churn significativamente mayores. Esto sugiere que los servicios complementarios ayudan a fidelizar.

* pago_automatico: Este método está asociado con menor churn. Posiblemente porque automatizar el pago ayuda a fidelizar al cliente.

4.- Facturación

* PaperlessBilling: Los clientes con facturación electrónica presentan ligeramente más churn. Esto podría estar correlacionado con perfiles más digitales, que también comparan y cambian más fácilmente de proveedor.

##Conclusiones estratégicas

- Invertir en servicios complementarios como respaldo, seguridad y soporte técnico ayuda a reducir la fuga de clientes.

- Los contratos largos y el pago automático son poderosos mecanismos de retención.

- Los clientes con menos vínculos personales o familiares (sin pareja ni dependientes) deberían ser segmentados para seguimiento personalizado.

- Si se detectan perfiles sin uso de servicios clave como internet o streaming, pueden estar en riesgo de abandono.

##Graficando variables numéricas

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches  # Para crear la leyenda manualmente

# Variables numéricas a graficar
variables =variables_numericas

# Paleta de colores personalizada
colores = {0: "#1f77b4", 1: "#ff7f0e"}  # Azul para No Churn, Naranjo para Churn

# Crear figura y ejes
fig, axs = plt.subplots(1, 2, figsize=(10,6))
axs = axs.flatten()

for ax, variable in zip(axs, variables):
    # Boxplot con color y media
    sns.boxplot(
        data=datos_depurados,
        x='Churn',
        y=variable,
        ax=ax,
        hue='Churn',
        palette=colores,
        dodge=False,
        showmeans=True,
        meanline=True,
        meanprops={"color": "red", "linestyle": "--", "linewidth": 2},
        legend=False  # Evita duplicación de leyenda
    )

    # Calcular medianas
    mediana_0 = datos_depurados[datos_depurados['Churn'] == 0][variable].median()
    mediana_1 = datos_depurados[datos_depurados['Churn'] == 1][variable].median()

    # Añadir texto de mediana
    ax.text(0, mediana_0, f'Mediana: {mediana_0:.2f}', ha='center', va='bottom', fontsize=10, color='black')
    ax.text(1, mediana_1, f'Mediana: {mediana_1:.2f}', ha='center', va='bottom', fontsize=10, color='black')

    # Título y etiquetas
    ax.set_xticklabels(['No Churn', 'Churn'])
    ax.set_title(f'Distribución de {variable} por Churn')

# Agregar leyenda manual (fuera del loop)
leyenda_patches = [
    mpatches.Patch(color=colores[0], label='No Churn (0)'),
    mpatches.Patch(color=colores[1], label='Churn (1)')
]

# Posicionar leyenda arriba a la derecha
fig.legend(
    handles=leyenda_patches,
    loc='upper right',
    title='Churn',
    fontsize=6,
    title_fontsize=6
)

# Mostrar y guardar
plt.tight_layout(rect=[0, 0, 0.95, 0.95])  # Ajustar espacio para la leyenda
plt.savefig("3_graficos_boxplot_variables_numericas.png", dpi=300)
plt.show()
```

##Interpretación del Mapa de Calor de correlaciones con Churn

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# Calculamos la correlación
correlaciones = datos_depurados.corr(numeric_only=True)

# Extraemos la fila de 'Churn' (sin incluir 'Churn' consigo mismo)
corr_churn = correlaciones.loc['Churn'].drop('Churn')

# Ordenamos por magnitud (valor absoluto)
corr_ordenadas = corr_churn.reindex(corr_churn.abs().sort_values(ascending=False).index)

# Clasificación del tipo de correlación
def clasificar_correlacion(valor):
    abs_val = abs(valor)
    if abs_val >= 0.3:
        return 'Alta'
    elif abs_val >= 0.2:
        return 'Media'
    elif abs_val >= 0.1:
        return 'Baja'
    else:
        return 'Insignificante'

# Interpretación contextual
def interpretar_correlacion(var, val):
    if val > 0:
        return f'A mayor {var}, mayor probabilidad de churn'
    elif val < 0:
        return f'A mayor {var}, menor probabilidad de churn'
    else:
        return f'{var} no tiene relación con churn'

# Construimos el DataFrame
tabla_corr = pd.DataFrame({
    'Variable': corr_ordenadas.index,
    'Correlación': corr_ordenadas.values,
    'Signo': ['+' if val > 0 else '–' for val in corr_ordenadas.values],
    'Tipo de Correlación': [clasificar_correlacion(val) for val in corr_ordenadas.values],
    'Interpretación': [interpretar_correlacion(var, val) for var, val in zip(corr_ordenadas.index, corr_ordenadas.values)]
})

# Crear figura tipo tabla con Plotly
fig = go.Figure(data=[go.Table(
    columnwidth=[70, 40, 40, 70, 200],
    header=dict(
        values=[
            'Variable',
            'Correlación<br><span style="font-size:11px">',
            'Signo<br><span style="font-size:11px">',
            'Tipo de Correlación<br><span style="font-size:11px">',
            'Interpretación<br><span style="font-size:11px">'
        ],
        fill_color='lightsteelblue',
        align='center',
        font=dict(size=14, color='black'),
        height=40
    ),
    cells=dict(
        values=[
            tabla_corr['Variable'],
            [f"{val:.2f}" for val in tabla_corr['Correlación']],
            tabla_corr['Signo'],
            tabla_corr['Tipo de Correlación'],
            tabla_corr['Interpretación']
        ],
        fill_color='white',
        align='center',
        font=dict(size=13),
        height=30
    )
)])

fig.update_layout(margin=dict(l=40, r=40, t=20, b=20))
fig.write_html("7_Interpretacion_Correlaciones_Churn.html")
fig.show()




```

##Preparación de datos

```python
import pandas as pd
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# División estratificada
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_idx, test_idx in split.split(datos_depurados, datos_depurados['Churn']):
    strat_train_set = datos_depurados.loc[train_idx]
    strat_test_set = datos_depurados.loc[test_idx]

# Separar variables independientes y dependiente (Churn) y Separación entrenamiento/prueba
X_train = strat_train_set.drop('Churn', axis=1)
y_train = strat_train_set['Churn']
X_test = strat_test_set.drop('Churn', axis=1)
y_test = strat_test_set['Churn']




```

##Entrenando el modelo

```python
import pandas as pd
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score,recall_score

# Modelo de regresión logística
modelo_lr = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
modelo_lr.fit(X_train, y_train)

# Evaluamos
y_pred_lr= modelo_lr.predict(X_test)

# Reporte de resultados
print("Matriz de confusión:")
print(confusion_matrix(y_test, y_pred_lr))

print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred_lr))

# Guardar valor de Recall como recall_lr
recall_lr = recall_score(y_test, y_pred_lr, average='binary')
print(f"\nValor de Recall guardado en recall_lr: {recall_lr:.2f}")

# Guardar valor del reporte como reporte_lr
reporte_lr = classification_report(y_test, y_pred_lr)
```
